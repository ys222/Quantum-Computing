# Quantum Bits (Qubits)

## Basics of Quantum Mechanics:

### 1. Quantum States:

In quantum mechanics, the state of a quantum system is represented by a state vector \(|\psi\rangle\), and its evolution is governed by the Schrödinger equation.

### 2. Superposition:

1. **Principle of Superposition:**
   - A key principle in quantum mechanics allowing quantum systems to exist in multiple states simultaneously.

2. **Mathematical Representation:**
   - A qubit in superposition is represented as \(|\psi\rangle = \alpha|0\rangle + \beta|1\rangle\), where \(\alpha\) and \(\beta\) are complex numbers and \(|0\rangle\) and \(|1\rangle\) are basis states.

3. **Bloch Sphere:**
   - A geometric representation of a qubit's state, where each point on the sphere corresponds to a unique qubit state.

## Representing Information Using Qubits:

### 1. Quantum Gates:

1. **Unitary Transformations:**
   - Quantum gates perform unitary transformations on qubits.

2. **Pauli-X, Y, Z Gates:**
   - Basic quantum gates that perform bit-flip and phase-flip operations.

### 2. Quantum Circuits:

1. **Quantum Circuit Model:**
   - Qubits are manipulated through quantum circuits composed of quantum gates.

2. **Quantum Measurement:**
   - Measurement collapses a qubit's superposition to one of its basis states with probabilities determined by the coefficients \(\alpha\) and \(\beta\).

## Entanglement and its Role:

### 1. Quantum Entanglement:

1. **Definition:**
   - A phenomenon where two or more qubits become correlated and the state of one qubit cannot be described independently of the others.

2. **Bell States:**
   - Examples of entangled states, such as \(|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)\).

### 2. Role in Quantum Information Processing:

1. **Quantum Teleportation:**
   - The transmission of quantum information using entangled particles.

2. **Quantum Cryptography:**
   - Utilizes entanglement for secure key distribution.

## Equations and Formulas:

### 1. Superposition:

1. **Superposition Equation:**
   \[ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle \]

### 2. Entanglement:

1. **Bell State Equation:**
   \[ |\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \]

### 3. Bloch Sphere Representation:

1. **Bloch Sphere Coordinates:**
   \[ |\psi\rangle = \cos\left(\frac{\theta}{2}\)|0\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\)|1\rangle \]

## Conclusion:

Quantum bits (qubits) leverage the principles of quantum mechanics, such as superposition and entanglement, to enable quantum information processing. Qubits, represented by state vectors, can exist in multiple states simultaneously. Quantum gates and circuits manipulate qubits, and entanglement plays a crucial role in quantum protocols like quantum teleportation and quantum cryptography. Understanding qubits and their properties is fundamental to the development of quantum algorithms and quantum computing.

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble

# Function to create a quantum circuit with superposition
def superposition_circuit():
    qc = QuantumCircuit(1)
    
    # Apply Hadamard gate for superposition
    qc.h(0)
    
    return qc

# Function to create a quantum circuit with entanglement
def entanglement_circuit():
    qc = QuantumCircuit(2)
    
    # Apply Hadamard gate on the first qubit
    qc.h(0)
    
    # Apply CNOT gate for entanglement
    qc.cx(0, 1)
    
    return qc

# Display quantum circuits
print("Superposition Circuit:")
superposition_circuit().draw('mpl')

print("\nEntanglement Circuit:")
entanglement_circuit().draw('mpl')
```
