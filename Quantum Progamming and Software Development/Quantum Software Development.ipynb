### Quantum Annealing:

Quantum annealing is a quantum computing paradigm designed to find the global minimum of a cost function. The system starts in a superposition of states and evolves toward the solution by minimizing the cost function. The annealing process is described by the Hamiltonian, and the time evolution is governed by the Schrödinger equation:

\[ i \hbar \frac{d}{dt} |\psi(t)\rangle = H(t) |\psi(t)\rangle \]

Where:
- \( \hbar \) is the reduced Planck constant,
- \( |\psi(t)\rangle \) is the state vector of the quantum system at time \( t \),
- \( H(t) \) is the time-dependent Hamiltonian.

The Hamiltonian in quantum annealing typically consists of a classical energy function and a tunneling term, representing quantum tunneling effects.

5. **Simulation:**

6. **Execution on Quantum Hardware:**

````python
from qiskit import QuantumCircuit, Aer, transpile, assemble

# Create a quantum circuit with 2 qubits
quantum_circuit = QuantumCircuit(2, 2)

# Add a Hadamard gate on qubit 0
quantum_circuit.h(0)

# Add a CNOT (Controlled-X) gate with qubit 0 as control and qubit 1 as target
quantum_circuit.cx(0, 1)

# Measure the qubits
quantum_circuit.measure([0, 1], [0, 1])

# Visualize the circuit
print("Quantum Circuit:")
print(quantum_circuit)

# Simulate the circuit using the Aer simulator
simulator = Aer.get_backend('qasm_simulator')

# Transpile the quantum circuit for the simulator
transpiled_circuit = transpile(quantum_circuit, simulator)

# Assemble the transpiled circuit for the simulator
qobj = assemble(transpiled_circuit)

# Run the simulation
result = simulator.run(qobj).result()

# Get and print the counts of the measurements
counts = result.get_counts()
print("\nMeasurement Counts:", counts)
````

# Introduction to Quantum Computing

## 1. **Quantum Bits (Qubits):**

### a. **Classical Bits vs. Qubits:**
   - Classical computers use bits as the smallest unit of information, which can be either 0 or 1.
   - Quantum computers use qubits, which can exist in a superposition of 0 and 1 simultaneously.

### b. **Superposition:**
   - Qubits can exist in a linear combination of states. If \(|0\rangle\) and \(|1\rangle\) are classical states, a qubit in superposition is represented as:
     \[ |\psi\rangle = \alpha|0\rangle + \beta|1\rangle \]
     where \(\alpha\) and \(\beta\) are complex numbers representing probability amplitudes.

### c. **Quantum Measurement:**
   - When a qubit is measured, it collapses to one of the classical states \(|0\rangle\) or \(|1\rangle\) with probabilities given by the squared magnitudes of the amplitudes.

## 2. **Superposition and Entanglement:**

### a. **Entanglement:**
   - Entanglement is a phenomenon where two or more qubits become correlated and the state of one qubit is dependent on the state of the other(s).
   - Example of an entangled state:
     \[ |\Psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \]

### b. **Bell States:**
   - Bell states are maximally entangled quantum states:
     \[ |\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \]
     \[ |\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle) \]
     \[ |\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle) \]
     \[ |\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle) \]

## 3. **Quantum Gates and Circuits:**

### a. **Quantum Gates:**
   - Quantum gates are analogous to classical logical gates but operate on qubits. Examples include Hadamard gate (H), Pauli gates (X, Y, Z), and CNOT gate.
   - Gates are represented by unitary matrices.

### b. **Quantum Circuit:**
   - A quantum circuit is a sequence of quantum gates applied to qubits.
   - Circuit evolution is described by the Schrödinger equation:
     \[ i\hbar \frac{\partial}{\partial t} |\psi(t)\rangle = H |\psi(t)\rangle \]

### c. **Quantum Entanglement using CNOT:**
   - The CNOT gate is commonly used to create entanglement between qubits. For example, applying CNOT to qubits \(q_1\) and \(q_2\) creates the entangled state:
     \[ |\Psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) \]

### d. **Quantum Superposition using Hadamard Gate:**
   - The Hadamard gate can create superposition. Applying Hadamard to a qubit initially in state \(|0\rangle\) results in:
     \[ H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \]

Quantum computing leverages the principles of superposition and entanglement to perform certain calculations exponentially faster than classical computers for specific problems. However, challenges like decoherence and error correction must be addressed for practical quantum computation. Quantum algorithms, such as Shor's algorithm and Grover's algorithm, showcase the power of quantum computing in factoring large numbers and searching unsorted databases. Ongoing research and advancements in quantum hardware and software aim to make quantum computing more robust and widely applicable.

````python
from qiskit import QuantumCircuit, Aer, transpile, assemble, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector

# Create a quantum circuit with 2 qubits
qc = QuantumCircuit(2)

# Apply Hadamard gate to create superposition
qc.h(0)

# Apply CNOT gate for entanglement
qc.cx(0, 1)

# Visualize the circuit
print("Quantum Circuit:")
print(qc)

# Simulate the quantum circuit
simulator = Aer.get_backend('statevector_simulator')
compiled_circuit = transpile(qc, simulator)
qobj = assemble(compiled_circuit)
result = simulator.run(qobj).result()

# Get and print the statevector
statevector = result.get_statevector()
print("\nStatevector:")
print(statevector)

# Plot Bloch vector for each qubit
plot_bloch_multivector(statevector)

# Measure the qubits
qc.measure_all()

# Visualize the compiled circuit after measurement
compiled_circuit = transpile(qc, simulator)
qobj = assemble(compiled_circuit)
result = simulator.run(qobj).result()

# Get and plot the measurement results
counts = result.get_counts()
print("\nMeasurement Results:")
print(counts)
plot_histogram(counts)
````

### 1. Shor's Algorithm (Integer Factorization):
- **Overview:**
  - Shor's algorithm is a quantum algorithm that efficiently factors large integers exponentially faster than the best-known classical algorithms.
  
- **Key Equations:**
  - The algorithm employs modular exponentiation and quantum Fourier transform techniques.
  - The period-finding algorithm is a critical component, where the quantum Fourier transform helps identify the period of a modular exponentiation function.

### 2. Grover's Algorithm (Database Search):
- **Overview:**
  - Grover's algorithm is designed to search an unsorted database using quantum parallelism, providing a quadratic speedup compared to classical algorithms.

- **Key Equations:**
  - The algorithm utilizes amplitude amplification to enhance the probability of finding the correct solution.
  - The number of iterations required for optimal search is proportional to the square root of the number of items in the database.

### 3. Quantum Fourier Transform (QFT):
- **Overview:**
  - The Quantum Fourier Transform is a fundamental operation in many quantum algorithms, including Shor's algorithm.
  
- **Key Equations:**
  - The Quantum Fourier Transform on \(n\) qubits is defined by the unitary transformation:
    \[ QFT |x\rangle = \frac{1}{\sqrt{2^n}} \sum_{y=0}^{2^n-1} e^{2\pi i \frac{xy}{2^n}} |y\rangle \]

### 4. Variational Quantum Algorithms:
- **Overview:**
  - Variational algorithms, such as the Variational Quantum Eigensolver (VQE), use parameterized quantum circuits to find approximate solutions to optimization problems.
  
- **Key Equations:**
  - In VQE, the goal is to find the set of parameters \(\theta\) that minimizes the expectation value of a given Hamiltonian \(H\):
    \[ \text{minimize} \langle \Psi(\theta) | H | \Psi(\theta) \rangle \]
  - The quantum state \(|\Psi(\theta)\rangle\) is prepared using a parameterized quantum circuit.

Please note that the equations provided here are simplified representations, and the actual implementation and mathematical details can be more involved. Quantum algorithms often involve linear algebra, complex numbers, and advanced mathematical concepts from quantum mechanics.

For more in-depth understanding and detailed equations, you may refer to specific literature, textbooks, or research papers on each algorithm. Each algorithm has a rich theoretical background, and their implementations can vary based on the specific quantum computing framework being used.

````python
from qiskit import Aer, QuantumCircuit, transpile, assemble
from qiskit.circuit.library import QFT

# Function to apply Quantum Fourier Transform
def quantum_fourier_transform(qubits):
    circuit = QuantumCircuit(len(qubits))
    circuit.h(qubits)
    circuit.append(QFT(len(qubits)), qubits)

    return circuit

# Example: Apply Quantum Fourier Transform on 3 qubits
qubits = [0, 1, 2]
qft_circuit = quantum_fourier_transform(qubits)
print("Quantum Fourier Transform Circuit:")
print(qft_circuit)
````

### Shor's Algorithm:
**Objective:** Factoring integers efficiently.

1. **Classical Factorization:**
   - Classical algorithms for factoring large numbers have exponential time complexity.
   
2. **Quantum Speedup:**
   - Shor's algorithm achieves exponential speedup over classical algorithms.
   
3. **Algorithm Steps:**
   - **Step 1:** Choose a random number \(a\) less than \(N\).
   - **Step 2:** Compute the greatest common divisor (GCD) of \(a\) and \(N\).
   - **Step 3:** If the GCD is not 1, that is a non-trivial factor.
   - **Step 4:** If the GCD is 1, use quantum period finding to find the order \(r\) of \(a\) modulo \(N\).
   - **Step 5:** If \(r\) is odd, go back to Step 1. If \(a^{r/2} \equiv -1 \pmod{N}\), go back to Step 1.
   - **Step 6:** The factors of \(N\) are then computed from \(a^{r/2} \pm 1\) using the GCD.

4. **Time Complexity:**
   - Quantum: \(O((\log N)^3)\)
   - Classical: \(O(e^{(\log N)^{1/3} (\log \log N)^{2/3}})\)

### Grover's Algorithm:
**Objective:** Searching an unsorted database quadratically faster than classical algorithms.

1. **Classical Search:**
   - Classical algorithms take \(O(N)\) time to search an unsorted database.

2. **Quantum Speedup:**
   - Grover's algorithm achieves a quadratic speedup, taking \(O(\sqrt{N})\) time.

3. **Algorithm Steps:**
   - **Step 1:** Prepare a uniform superposition of all possible states.
   - **Step 2:** Apply an oracle to mark the target state.
   - **Step 3:** Amplify the amplitude of the target state.
   - **Step 4:** Repeat Steps 2 and 3 about \(\sqrt{N}\) times.
   - **Step 5:** Measure the state, obtaining the solution with high probability.

4. **Time Complexity:**
   - Quantum: \(O(\sqrt{N})\)
   - Classical: \(O(N)\)

### Quantum Fourier Transform (QFT):
**Objective:** Efficiently perform Fourier transform on quantum data.

1. **Classical Fourier Transform:**
   - \(X_k = \sum_{n=0}^{N-1} x_n \cdot e^{-i 2\pi \frac{kn}{N}}\)
   
2. **Quantum Speedup:**
   - QFT provides an exponential speedup over classical Fourier transforms.

3. **QFT Equation:**
   - \(QFT \ket{j} = \frac{1}{\sqrt{N}} \sum_{k=0}^{N-1} e^{2\pi i \frac{jk}{N}} \ket{k}\)

4. **QFT Implementation:**
   - Involves Hadamard gates and controlled-phase gates.

### Variational Quantum Algorithms:
**Objective:** Solve optimization problems using parameterized quantum circuits.

1. **Basic Idea:**
   - Represent a problem with a cost function that quantum circuits try to minimize.

2. **Variational Form:**
   - \(|\psi(\theta)\rangle = U(\theta) |0\rangle\) where \(U(\theta)\) is a parameterized quantum circuit.

3. **Variational Quantum Eigensolver (VQE):**
   - Used for finding the ground state energy of a quantum system.
   - \(E(\theta) = \frac{\langle\psi(\theta)|H|\psi(\theta)\rangle}{\langle\psi(\theta)|\psi(\theta)\rangle}\)

4. **Quantum Circuit Layers:**
   - Alternating layers of single-qubit rotations and entangling gates.

5. **Optimization Process:**
   - Use classical optimization algorithms to adjust parameters for minimum cost.

````python
"""
This module contains the Quantum Fourier Transform class.
"""

from numpy import pi
from qiskit import QuantumCircuit, Aer, transpile, assemble


class QuantumFourierTransform:
    """Class for operations of the Quantum Fourier Transform."""

    @staticmethod
    def simulate(state: int) -> dict:
        """Simulate the QFT and iQFT.

        Parameters
        ----------
            state (int): The state to simulate.

        Returns
        -------
            dict: The simulated state.
        """
        # Get the number of qubits.
        qubit_count = state.bit_length()

        # Create the circuit.
        circuit = QuantumCircuit(qubit_count, qubit_count)

        # Apply the initial state.
        for qubit in range(qubit_count):
            if state & (1 << qubit):
                circuit.x(qubit)

        # Apply the QFT.
        circuit = QuantumFourierTransform.qft(circuit)

        # Apply the inverse QFT.
        circuit = QuantumFourierTransform.iqft(circuit)

        # Append the measurement.
        circuit.measure(range(qubit_count), range(qubit_count))

        # Run the simulation.
        simulator = Aer.get_backend("aer_simulator")
        circuit = transpile(circuit, simulator)
        job = simulator.run(assemble(circuit))
        result = job.result().get_counts()

        answer_as_list = list(result.keys())
        answer_int = int(answer_as_list[0], 2)
        return {"result": answer_int}

    @staticmethod
    def qft(circuit: QuantumCircuit) -> QuantumCircuit:
        """Apply QFT to a circuit.

        Parameters
        ----------
            circuit (QuantumCircuit):  The circuit to apply the QFT to.

        Returns
        -------
            QuantumCircuit: The circuit with the QFT applied.
        """
        # Apply the QFT to the circuit.
        circuit = QuantumFourierTransform._qft_append_circuit(
            circuit, circuit.num_qubits - 1
        )

        # Apply the swaps to the circuit.
        circuit = QuantumFourierTransform._qft_append_swaps(circuit)

        return circuit

    @staticmethod
    def qft_circuit(qubit_count: int) -> QuantumCircuit:
        """Create a QFT circuit with given Qubit count.

        Parameters
        ----------
        qubit_count : int
            The number of qubits to use in the circuit.

        Returns
        -------
        QuantumCircuit
            The QFT circuit.
        """
        return QuantumFourierTransform.qft(QuantumCircuit(qubit_count))

    @staticmethod
    def iqft_circuit(qubit_count: int) -> QuantumCircuit:
        """Create a iQFT circuit with given Qubit count.

        Parameters
        ----------
        qubit_count : int
            The number of qubits to use in the circuit.

        Returns
        -------
        QuantumCircuit
            The iQFT circuit.
        """
        return QuantumFourierTransform.iqft(QuantumCircuit(qubit_count))

    @staticmethod
    def iqft(circuit: QuantumCircuit) -> QuantumCircuit:
        """Apply inverse QFT to a circuit.

        Parameters
        ----------
            circuit (QuantumCircuit):  The circuit to apply the IQFT to.

        Returns
        -------
            QuantumCircuit: The circuit with the iQFT applied.
        """
        # Apply the swaps to the circuit.
        circuit = QuantumFourierTransform._qft_append_swaps(circuit, inverse=True)
        circuit.barrier()

        # Apply the QFT to the circuit.
        circuit = QuantumFourierTransform._iqft_append_circuit(circuit, 0)

        return circuit

    @staticmethod
    def _qft_append_circuit(circuit: QuantumCircuit, qubit: int) -> QuantumCircuit:
        """Apply a rotation to a qubit.

        Parameters
        ----------
            circuit (QuantumCircuit): The circuit to apply the rotation to.
            qubit (int): The qubit to apply the rotation to.

        Returns
        -------
            QuantumCircuit: The circuit with the rotation applied.
        """
        # Recursive stop condition.
        if qubit < 0:
            return circuit

        # Construct the minimal QFT circuit.
        circuit.h(qubit)
        for qubit_line in reversed(range(qubit)):
            circuit.cp(pi / 2 ** (qubit - qubit_line), qubit_line, qubit)
        circuit.barrier()

        # Recursively apply the QFT to the next qubit.
        return QuantumFourierTransform._qft_append_circuit(circuit, qubit - 1)

    @staticmethod
    def _iqft_append_circuit(circuit: QuantumCircuit, qubit: int) -> QuantumCircuit:
        """Apply a rotation to a qubit.

        Parameters
        ----------
            circuit (QuantumCircuit): The circuit to apply the rotation to.
            qubit (int): The qubit to apply the rotation to.

        Returns
        -------
            QuantumCircuit: The circuit with the rotation applied.
        """
        # Recursive stop condition.
        if qubit >= circuit.num_qubits:
            return circuit

        # Construct the minimal QFT circuit.
        for qubit_line in range(qubit):
            circuit.cp(-pi / 2 ** (qubit - qubit_line), qubit_line, qubit)
        circuit.h(qubit)
        circuit.barrier()

        # Recursively apply the QFT to the next qubit.
        return QuantumFourierTransform._iqft_append_circuit(circuit, qubit + 1)

    @staticmethod
    def _qft_append_swaps(
        circuit: QuantumCircuit, inverse: bool = False
    ) -> QuantumCircuit:
        """Apply swaps to a circuit.

        Parameters
        ----------
            circuit (QuantumCircuit): The circuit to apply the swaps to.
            qubit (int): The qubit to apply the rotation to.

        Returns
        -------
            QuantumCircuit: The circuit with the swaps applied.
        """
        qubit_count = circuit.num_qubits
        qubits_list = (
            reversed(range(qubit_count // 2))
            if not inverse
            else range(qubit_count // 2)
        )
        for qubit in qubits_list:
            circuit.swap(qubit, qubit_count - qubit - 1)
        return circuit



if __name__ == "__main__":
    print("===================================")
    print("Quantum Fourier Transform Simulator")
    print("===================================")

    # Get the input state as integer decimal.
    state_int = int(input("> Enter the state as decimal integer: "))

    # Run the algorithm.
    result = QuantumFourierTransform.simulate(state_int)
    print(f"iQFT result: {result['result']}")
````
